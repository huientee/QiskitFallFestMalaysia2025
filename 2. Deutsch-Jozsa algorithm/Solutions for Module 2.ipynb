```python
from qiskit import QuantumCircuit

def oracle2():

    oracle2 = QuantumCircuit(5)

    return oracle2
```

```python
# Import required modules
from qiskit import QuantumCircuit
from qiskit_aer.primitives import SamplerV2 as AerSampler
from qiskit.visualization import plot_histogram

# Number of input qubits
n = 4  # match the given oracle

# Create a circuit: n input qubits + 1 output qubit
qc_dj = QuantumCircuit(n + 1, n)

# Set the output qubit to |1>
qc_dj.x(n)

# Apply Hadamard gates to all qubits
qc_dj.h(range(n + 1))

# Add oracle
qc_dj.barrier()
qc_dj.compose(oracle2(), inplace=True)
qc_dj.barrier()

# Apply Hadamard gates to all input qubits
qc_dj.h(range(n))

# Measure all input qubits
qc_dj.measure(range(n), range(n))

# Draw the circuit diagram
qc_dj.draw("mpl")

# Run the job on simulator
sampler_sim = AerSampler()
job = sampler_sim.run([qc_dj], shots=1000)
res = job.result()

# Analyse results
counts = res[0].data.c.get_counts()

print(counts)
if (
    "0" * n in counts
):  # The algorithm returns all zeroes if the function is constant
    print("constant")
else:
    print("balanced")  # anything other than all zeroes means the function is balanced
```
