<h3><b> Description </b></h3>

Bloch sphere geometrically represents the state of a single qubit. Each point on the surface of the sphere corresponds to a possible qubit state.

A general qubit state can be written as

$$
\lvert\psi\rangle = \cos\!\frac{\theta}{2}\,\lvert 0\rangle \;+\; e^{i\phi}\sin\!\frac{\theta}{2}\,\lvert 1\rangle
$$

where $0 \leq \theta \leq \pi$ and $0 \leq \phi < 2\pi$. In this representation:

- North pole ($\theta=0$) is the state $\lvert0\rangle$, and south pole ($\theta=\pi$) is the state $\lvert1\rangle$.
- Angle $\theta$ (polar angle) measures how far the state is from $\lvert0\rangle$, changing $\theta$ moves the point up and down between the poles.
- Angle $\phi$ (azimuthal angle) is the relative phase between $\lvert0\rangle$ and $\lvert1\rangle$, changing $\phi$ rotates the state around the $z$-axis of the Bloch sphere.

Note: points on the equator ($\theta=\tfrac{\pi}{2}$) are equal superpositions of $\lvert0\rangle$ and $\lvert1\rangle$:
  $$
  \phi = 0 \;\;\Rightarrow\;\; \tfrac{\lvert0\rangle + \lvert1\rangle}{\sqrt{2}}, \qquad
  \phi = \pi \;\;\Rightarrow\;\; \tfrac{\lvert0\rangle - \lvert1\rangle}{\sqrt{2}}.
  $$

  <hr>

<h3><b> Visualisation of the state </b></h3>

To visualize a qubit state on the Bloch sphere, the state can be described using either **Cartesian** or **spherical** coordinates.

**Cartesian coordinates** $(x, y, z)$ of a qubit state are related to the Bloch sphere angles $(\theta, \phi)$ by:

- $$x = \sin\theta \cos\phi$$
- $$y = \sin\theta \sin\phi$$
- $$z = \cos\theta$$

In **spherical coordinates**, a point on the Bloch sphere is described by three parameters $(r, \theta, \phi)$.

Since pure qubit states correspond to points **on the surface** of the sphere, the radius is fixed at $r = 1$ (unit sphere). Therefore, only the angles $\theta$ and $\phi$ are required to specify a qubit state.

<h3><b> Example 1: Using Cartesian coordinates </b></h3>

Given a qubit state  
$$
\lvert\psi\rangle = \cos\!\frac{\pi}{4}\lvert0\rangle + e^{i\pi/3}\sin\!\frac{\pi}{4}\lvert1\rangle,
$$
what does it look like on the Bloch sphere?

```python
!pip install qiskit qiskit[visualization]
```

```python
import numpy as np
from qiskit.visualization import plot_bloch_vector

# Given theta = π/2, phi = π/3
theta = np.pi / 2
phi = np.pi / 3

# Convert spherical → cartesian for plotting
x = np.sin(theta) * np.cos(phi)
y = np.sin(theta) * np.sin(phi)
z = np.cos(theta)

plot_bloch_vector([x, y, z], title="First Bloch Sphere")
```

<h3><b> Example 2: Using spherical coordinates </b></h3>

Use the same qubit state from **Example 1**, and plot it on the Bloch sphere by specifying its **spherical coordinates** directly.

Here, we set $r = 1$, $\theta = \frac{\pi}{2}$, and $\phi = \frac{\pi}{3}$.


```python
import numpy as np
from qiskit.visualization import plot_bloch_vector

# Plot the same qubit using spherical coordinates
plot_bloch_vector([1, np.pi/2, np.pi/3], coord_type='spherical', title="Second Bloch Sphere")
```
